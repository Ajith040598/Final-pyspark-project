```
df2=spark.read.json('/FileStore/tables/yelp_academic_dataset_business.json')
```

```
df2.write.option("header","true").csv("business1.csv")
df2.printSchema()
```

```
df2=df2.withColumn("categories",explode(split("categories",",")))
df2=df2.withColumn("attributes",col("attributes").cast("string"))
df2=df2.withColumn("hours",col("hours").cast("string"))
df2.coalesce(1).write.format("com.databricks.spark.csv").option("header","true").save("business3")
df2.write.option("header","true").csv("business3.csv")
df2.display()
```

```
df.createOrReplaceTempView("df")
df1.createOrReplaceTempView("df1")
df2.createOrReplaceTempView("df2")
```

```
data_spark=spark.sql("select name,categories,review_count,stars from df2 where categories='Restaurants' and review_count>=10 order By stars desc,review_count desc limit 5 ")
display(data_spark)
```

```
data_spark1=spark.sql("select business_id,name,categories,review_count,stars from df2 where categories='Home Services' order By review_count desc limit 5")
display(data_spark1)
```

```
data_spark2=spark.sql("select city,max(stars) as highest_rated_nightlife from df2 where categories like '%Nightlife%' group By stars,city order By stars desc limit 1")
display(data_spark2)
```

```
df3=spark.sql("select df1.user_id,df2.name,df2.categories,df2.review_count from df1 inner join df2 on df1.business_id=df2.business_id")
df3.display()
df3.write.option("header","true").csv("users data.csv")
```
